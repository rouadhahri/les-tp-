```python


def dicho(a,b,f,tol):
    m=(a+b)/2
    err=abs(b-a)
    while err>tol:
        if f(m)==0 :
            break
        elif f(a)*f(m)<0 :
            b=m
        else :
            a=m
        m=(a+b)/2
        err=abs(b-a)
    return m
```

```python
def dichow (a ,b,f,eps):

    g,d = min (a,b),max(a,b)
    gau,dor = f(g),f(d)
    n = 0
    while d-g > eps :
        n += 1
        m = ( g + d )/2
        valm = f ( m )
        if gau * valm < 0 :
            d = m
            dor = valm
        else :
            g = m
            gau = valm
    
    return (g+d)/2,n
```

```python
def pt_fixe(f,x_0,e,n):
    a=x_0
    for i in range(n):
        x=a
        a=f(x)
        err=abs(a-x)
        print(a)
        if err<e:
            return a,i
    return "Divergence :( "
```

```python
def newtonn(f,fprime,a,eps):
    x1=a
    x2=x1-f(x1)/fprime(x1)
    while abs(x1-x2)>eps :
        x1=x2
        x2=x1-f(x1)/fprime(x1)
    return x2
```

```python
import math
from sympy import *
from sympy.abc import x

f=lambda x: math.exp(-x)-x
fprime=lambda x: -math.exp(-x)-1
newtonn(f,fprime,0,0.001)
```
